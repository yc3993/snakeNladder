```python
import matplotlib.pyplot as plt
```

```python
x = [10, 100, 1000, 10000, 100000]
y1 = [423, 711, 1897, 7081, 52219]
y2 = [499, 997, 3864, 21110, 183815]
```

```python
plt.plot(x,y1,label="Save")
plt.plot(x,y2, label="Reload")
plt.xlabel("Number of Snakes&Ladders")
plt.ylabel("Time/ns")
plt.title("Save&Reload Time")
plt.annotate("Number of Players = 10", xy=(50000, 175000))
plt.legend()
```
